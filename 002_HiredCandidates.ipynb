```python
import os
import pandas as pd
from sqlalchemy import create_engine
from dotenv import load_dotenv
import matplotlib.pyplot as plt
```

```python
load_dotenv("../.env")

user = os.getenv("PG_USER")
password = os.getenv("PG_PASSWORD")
host = os.getenv("PG_HOST")
port = os.getenv("PG_PORT")
database = os.getenv("PG_DATABASE")
```

```python
url = f"postgresql://{user}:{password}@{host}:{port}/{database}"
engine = create_engine(url)
```

```python
query = 'SELECT * FROM "Candidates";'

df = pd.read_sql(query, engine)

df.head()
```

Este código creará una nueva columna hired en su DataFrame, que será 1 si el candidato obtuvo una puntuación de 7 o más tanto en la Entrevista Técnica como en el Desafío de Codificación. En caso contrario, será 0.

```python
df['Hired'] = ((df['Technical Interview Score'] >= 7) & (df['Code Challenge Score'] >= 7)).astype(int)

df.head()
```

Este codigo generara una nueva colunma Category of Techonoly que busca agrupar la colunma Technology en categorias

```python
technology_category = {
    'Development - CMS Backend': 'Development',
    'Development - FullStack': 'Development',
    'Development - Frontend': 'Development',
    'Development - Backend': 'Development',
    'Game Development': 'Development',
    'Adobe Experience Manager': 'Development',
    'Development - CMS Frontend': 'Development',
    'QA Automation': 'Quality Assurance',
    'QA Manual': 'Quality Assurance',
    'DevOps': 'Operations',
    'System Administration': 'Operations',
    'Database Administration': 'Data',
    'Data Engineer': 'Data',
    'Business Intelligence': 'Data',
    'Salesforce': 'Sales and Business',
    'Sales': 'Sales and Business',
    'Client Success': 'Sales and Business',
    'Business Analytics / Project Management': 'Sales and Business',
    'Social Media Community Management': 'Marketing and Communication',
    'Mulesoft': 'Marketing and Communication',
    'Technical Writing': 'Marketing and Communication',
    'Security Compliance': 'Security',
    'Security': 'Security',
    'Design': 'Design',
}

df["Category of Techonoly"] = df["Technology"].map(technology_category) 
```

```python
df["Application Date"] = pd.to_datetime(df["Application Date"])
```
